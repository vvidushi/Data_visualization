```python
import pandas as pd
import os
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')  # Suppress warnings for cleaner output

# File paths
base_path = "/Users/vidushi/Documents/Data_Visualization/datasets/olympics"
files = ["teams.csv", "medallists.csv", "medals.csv", "medals_total.csv", 
         "athletes.csv", "coaches.csv", "events.csv"]

# Function to safely read CSV files
def safe_read_csv(file_path):
    try:
        return pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"Warning: File not found: {file_path}")
        return None

# Read files dynamically into a dictionary of DataFrames
dataframes = {}
for file in files:
    file_path = os.path.join(base_path, file)
    df = safe_read_csv(file_path)
    if df is not None:
        dataframes[file.replace('.csv', '')] = df

# Display the number of files successfully loaded
print(f"Successfully loaded {len(dataframes)} out of {len(files)} files.")

# Display basic information about each DataFrame
for name, df in dataframes.items():
    print(f"\n{name} shape: {df.shape}")
    print(f"{name} columns: {df.columns.tolist()}")
    display(df.head(2))  # Show first 2 rows of each DataFrame
```

# Extract the list of athletes_codes from missed_teams_event and flatten it into a single list


```python
import pandas as pd
import ast

# Convert string representations of lists to actual lists (skip invalid entries)
def safe_literal_eval(value):
    try:
        # Check if the string starts with '[' to identify list-like entries
        if isinstance(value, str) and value.strip().startswith('['):
            return ast.literal_eval(value)
        else:
            return None  # Skip non-list entries (e.g., "Unknown Code")
    except:
        return None

# Apply the function to the column
missed_teams_event['athletes_codes'] = missed_teams_event['athletes_codes'].apply(safe_literal_eval)

# Flatten the list and remove duplicates
flattened_list = list(set(
    item for sublist in missed_teams_event['athletes_codes'].dropna() 
    for item in sublist
))

print("Distinct athletes_codes:")
print(flattened_list)
```

```python
filtered_list = [code for code in flattened_list if not code.startswith(('A', 'T'))]
print(filtered_list)
```

```python
# Check which codes in filtered_list exist in athletes_df['code']
matched_codes = [code for code in filtered_list if code in athletes_df['code'].values]

# Display matched codes
print("Matched Codes:")
print(matched_codes)
```

```python
# Fetch rows from athletes_df where the code matches the values in matched_codes
matched_athletes = athletes_df[athletes_df['code'].isin(matched_codes)]
matched_events_and_codes = matched_athletes[['events', 'code']]

matched_events_and_codes.head(5)
```
