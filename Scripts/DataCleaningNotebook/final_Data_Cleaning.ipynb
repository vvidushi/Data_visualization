# Extract the list of athletes_codes from missed_teams_event and flatten it into a single list


```python
import pandas as pd
import ast

# Convert string representations of lists to actual lists (skip invalid entries)
def safe_literal_eval(value):
    try:
        # Check if the string starts with '[' to identify list-like entries
        if isinstance(value, str) and value.strip().startswith('['):
            return ast.literal_eval(value)
        else:
            return None  # Skip non-list entries (e.g., "Unknown Code")
    except:
        return None

# Apply the function to the column
missed_teams_event['athletes_codes'] = missed_teams_event['athletes_codes'].apply(safe_literal_eval)

# Flatten the list and remove duplicates
flattened_list = list(set(
    item for sublist in missed_teams_event['athletes_codes'].dropna() 
    for item in sublist
))

print("Distinct athletes_codes:")
print(flattened_list)
```

```python
filtered_list = [code for code in flattened_list if not code.startswith(('A', 'T'))]
print(filtered_list)
```

```python
teams_df_cleaned = teams_df.dropna(subset=['athletes', 'athletes_codes', 'num_athletes'])
teams_df_cleaned['events'] = teams_df_cleaned['events'].fillna('UNKNOWN')
print(f"Original DataFrame shape: {teams_df.shape}")
print(f"Cleaned DataFrame shape: {teams_df_cleaned.shape}")
print("\nMissing values after cleaning:")
print(teams_df_cleaned[['athletes', 'athletes_codes', 'num_athletes', 'events']].isnull().sum())
cleaned_file_path = "/Users/vidushi/Documents/Data_Visualization/datasets/olympics/cleaned_teams.csv"
teams_df_cleaned.to_csv(cleaned_file_path, index=False)
print(f"Cleaned data saved to: {cleaned_file_path}")

```

Udf compare_dataframes

```python

def compare_dataframes(original_df, cleaned_df):
    print("Columns in original but not in cleaned:")
    print(set(original_df.columns) - set(cleaned_df.columns))
    print("\nColumns in cleaned but not in original:")
    print(set(cleaned_df.columns) - set(original_df.columns))
    print(f"\nOriginal shape: {original_df.shape}")
    print(f"Cleaned shape: {cleaned_df.shape}")
    print("\nMissing values in original:")
    print(original_df.isnull().sum())
    print("\nMissing values in cleaned:")
    print(cleaned_df.isnull().sum())
    print(f"Duplicates in cleaned: {cleaned_df.duplicated().sum()}")
    # print("\nRows in original but not in cleaned:")
    diff1 = pd.concat([original_df, cleaned_df, cleaned_df]).drop_duplicates(keep=False)
    print(diff1.head())
    print("\nRows in cleaned but not in original:")
    diff2 = pd.concat([cleaned_df, original_df, original_df]).drop_duplicates(keep=False)
    print(diff2.head())


```
