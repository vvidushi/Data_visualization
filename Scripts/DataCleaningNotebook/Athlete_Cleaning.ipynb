```python
# We dont need multiple columns for analysis and dropping the columns 
athlete_data= athlete_data.drop(['birth_place', 'birth_country', 'residence_place', 'residence_country', 'nickname','hobbies'], axis=1)
```

```python
athlete_data = athlete_data.drop(['occupation','education','family','lang'], axis=1)
```

```python
athlete_data.info()
```

```python
# Some of the columns like reason, hero, influence, philosophy, sporting_relatives, ritual etc.
athlete_data = athlete_data.drop(['reason','hero','influence','philosophy', 'sporting_relatives','ritual'], axis=1)
```

```python
athlete_data.info()
```

```python
athlete_data
```

```python
# By exploring the data we can see that there is no value in height its all 0, we can drop it
athlete_data = athlete_data.drop(['weight'], axis=1)
```

```python
athlete_data
```

```python
# Format the discliplies 
athlete_data['disciplines'] = athlete_data['disciplines'].astype(str).str.strip("[]").str.replace("'", "").str.replace('"', '')
```

```python
athlete_data
```

```python
athlete_data['events'] = athlete_data['events'].astype(str).str.strip("[]").str.replace("'", "").str.replace('"', '')
```

```python
athlete_data
```

```python
# Convert the 'birth_date' column to datetime
athlete_data['birth_date'] = pd.to_datetime(athlete_data['birth_date'])
```

```python
athlete_data.info()
```

```python
# Calculate age of the athletes by substracting the current date to birth date 

# Reference date (e.g., start of Paris 2024 Olympics: July 26, 2024)
reference_date = pd.to_datetime('2024-07-26')
# Calculate age
athlete_data['age'] = (reference_date - athlete_data['birth_date']).dt.days // 365
```

```python
athlete_data
```

```python
# We dont need 'coach' and 'other_sports'
```

```python
athlete_data = athlete_data.drop(['coach','other_sports'], axis=1)
```

```python
athlete_data
```

```python
athlete_data.info()
```

```python
athlete_data[athlete_data['height'].isnull()]
```

```python
# Replace NAN in '0'
athlete_data['height'] = athlete_data['height'].fillna('0')
```

```python
athlete_data[athlete_data['height'].isnull()]
```

```python
# Formatting in Name columns
def format_name(name):
    parts = name.split()
    if len(parts) == 2:
        last, first = parts
        return f"{first.capitalize()} {last.capitalize()}"
    else:
        return name.title()  # fallback for unexpected formats

# Apply to the name column
athlete_data['name'] = athlete_data['name'].apply(format_name)
```

```python
athlete_data
```

```python
#Rename columns 
athlete_data = athlete_data.rename(columns={
    'code':'Code',
    'current':'Current',
    
    'name': 'Name',
    
    'gender': 'Gender',
    'function':'Function',
    'country_code': 'Country Code',
    'country': 'Country',
    'nationality':'Nationality',
    'height':'Height',
     'disciplines': 'Discipline',
    'events':'Events',
    'birth_date': 'Birth Date',
    'age': 'Age'
   
})
```

```python
athlete_data
```

```python
athlete_data.to_csv('athlete_data.csv', index=False)
```
